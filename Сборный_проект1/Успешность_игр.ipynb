## Цель проекта
Необходимо определить параметры, которые приводят к росту популярности игр. Будут оцениваться какие игровые платформы наиболее популярны, как на продажу игры вдияют отзывы, какие жанры имеют наибольшую популярность и как меняется уровень продаж в заивисимости от региона. Анализ будет производится на выборке данных включающих информацию с 1980 по 2016 год.
## Подготовка данных

```python
data = pd.read_csv('/datasets/games.csv')
display(data.head(10))
display(data.describe())
data.info()
```

**Выводы и план предобработки:**
1. запишем названия столбцов с использованием только нижнего регистра;
2. заменим тип данных в столбце Year_of_Release на целочисленный;
3. разберёмся, почему в User_Score данные имеют тип object, хотя по аналогии с Critic_Score они должны иметь числовой формат;
4. заполним пропуски в столбцах:
    - Name: всего две строки с пропущенным именем игры, удаление этих строк не исказит статистику;
    - Year_of_Release: по названию игры и другим характеристикам можно установить год выхода, но пропусков слишком много для ручного заполнения. Может быть получится заполнить пропуски используя имя игры;
    - Genre: по названию игры и другим характеристикам можно восстановить жанр;
    - Critic_Score, User_Score: возможно, пропуски обусловлены отсутствием отзывов, тогда их можно заполнить нулями; 
    - Rating: пока непонятно, почему значения рейтинга могут отсутствовать, в зависимости от причины можно будет заполнить пропуски или медианным значением или чем-то еще;

```python
# изменим названия столбцов:
data.columns= data.columns.str.lower()

# удалим строки с пропусками в названии игр:
display(data.query('name.isna()'))
data = data.dropna(subset=['name']).reset_index(drop=True)
```

```python
# заполним пропуски в годах выхода, где возможно:
names_list = data.query('year_of_release.isna()')['name'].unique().tolist()
new_year = []
for name in names_list:
    flag = False
    for word in name.split(): 
        if '200' in word or '201' in word:
            flag = True
            year = word
    if flag:
        new_year.append([name, year])
print(len(new_year))
```

Удаление строк с пропусками в именах также решило проблему пропусков в жанрах. Пропусков в столбце year_of_release слишком много, чтобы заполнить их вручную, но при этом не слишком много, чтобы сильно повлиятть на статистику -- 2%; к сожалению, игр, содержащих год в названии всего 14 шт. Оставим пропуски как есть и изменим тип колонки на целочисленный.

```python
data['year_of_release'] = data['year_of_release'].astype('Int64')
```

```python
# посмотрим на пропуски в оценках:
print('Оценки критиков:\n', data['critic_score'].value_counts().head())
print('Оценки пользователей:\n', data['user_score'].value_counts().head())
```

У пользователей есть оценка tbd. TBD можно расшифровать как 'to be determined', значит, эти игры еще не успели оценить. Заполним пропуски в обеих колонках 0.

```python
data['critic_score'] = data['critic_score'].fillna(0)
data['user_score'] = data['user_score'].fillna(0)
data.loc[data['user_score'] == 'tbd', 'user_score'] = 0
data['user_score'] = data['user_score'].astype(float)
```

```python
# посмотрим, что там с пропусками в рейтинге:
data['rating'] = data['rating'].str.lower()
print(data.query('rating.isna()')['user_score'].value_counts().head())
print(data.query('rating.isna()')['critic_score'].value_counts().head())
display(data.query('rating.isna()'))
display(data['rating'].value_counts())
print(data.query('rating.isna()')['year_of_release'].value_counts())
```

ESRB-рейтинг появился в 1994 году, в нашем датасете есть игры 1980-1993 годов выхода, так что не удивительно, что для них не указан рейтинг. Оставшаяся (большая) часть пропущенного рейтинга относится к играм без оценки пользователей и критиков, возможно, для этих игр сломалось что-то при выгрузке данных.. Заполним пропуски значением unknown.

```python
data['rating'] = data['rating'].fillna('unknown')
display(data['rating'].value_counts())
```

## Исследовательский анализ данных
### Исследование зависимости количества продаж от типа  игровой плтаформы
#### Выбор актуального временного диапазона

```python
years_set = data.pivot_table(index='year_of_release', values='name', aggfunc='count').reset_index()
ticks_list = []
for i in range(1980, 2020, 5):
    ticks_list.append(i)

plt.figure(figsize=(20,20))
s = sns.catplot(x='year_of_release', y='name',data=years_set, kind='bar', ci=None, estimator=sum, color='seagreen')
s.set(xticks=np.arange(0,40,5), xlabel='Год выпуска игры', ylabel='Количество игр выпущенных за год',
      title='Распределение количества выпущенных игр по годам')
s.set_xticklabels(ticks_list, rotation = 90)
plt.show()
```

Распределение имеет вид нормального, наибольшее количество игр было выпущено между 2008-2009 годами. Будем рассматривать только игры выпущенные после 1995 года, так до этого момента количество игр в разы меньше.

#### Изменение количества продаж в зависимости от платформы

```python
platform_set = data.query('year_of_release >= 1995').pivot_table(index=['platform', 'year_of_release'], 
                                                                        values='sales_sum',
                                                                       aggfunc='sum').reset_index()

order = platform_set.groupby('platform').sum().sort_values(by='sales_sum', ascending=False).index

plt.figure(figsize=(8,19))
s = sns.catplot(x='platform', y='sales_sum', data=platform_set, kind='bar', ci=None, 
                estimator=sum, color='seagreen', order=order)
s.set_xticklabels(rotation = 90)
s.set(title='Изменение количества продаж в зависимости от платформы', 
      xlabel='Платформа', ylabel='Количество продаж, млн. копий')

plt.show()
```

Наибольшие суммарные продажи имеют игры для следующих платформ:
* PS2;
* X360;
* PS3;
* Wii;
* DS;

### Проверка гипотез
#### Сравнение пользовательских рейтингов различных платформ
Необходимо проверить следующую гипотезу: средние пользовательские рейтинги платформ Xbox One и PC одинаковые. Сформулируем нулевую и альтернативную гипотезу следующим образом:
* H$_{0}$: средние пользовательские рейтинги платформ Xbox One и PC одинаковые;
* H$_{1}$: средние пользовательские рейтинги платформ Xbox One и PC отличаются;

Сравнение будем проводить с помощью метода <code>ttest_ind (array1, array2, equal_var)</code>; для того, чтобы определить значение переменной equal_var воспользуемся тестом Барлетта:

```python
xbox_set = data_actual.query('platform=="XOne"')
pc_set = data_actual.query('platform=="PC"')
p = st.bartlett(xbox_set['user_score'], pc_set['user_score'])
print(np.where(p.pvalue < 0.05, 'Выборки значительно различаются: equal_var=False', 
         'Выборки различаются незначительно: equal_var=True'))
```

```python
result = st.ttest_ind(xbox_set['user_score'], pc_set['user_score'])
alpha = 0.05
print('p-value: ', result.pvalue)
print(np.where(result.pvalue < alpha, 'Отвергаем нулевую гипотезу: пользовательские рейтинги отличаются', 
         'Принимаем нулевую гипотезу: рейтинги практически одинаковые'))

mean_xbox = xbox_set['user_score'].mean()
mean_pc = pc_set['user_score'].mean()
ste_xbox = np.std(xbox_set['user_score'], ddof=1)/np.sqrt(len(xbox_set['user_score']))
ste_pc = np.std(pc_set['user_score'], ddof=1)/np.sqrt(len(pc_set['user_score']))
print('Средний рейтинг для платформы Xbox One: {0:.2f}\u00B1{1:.2f}'.format(mean_xbox, ste_xbox))
print('Средний рейтинг для платформы PC: {0:.2f}\u00B1{1:.2f}'.format(mean_pc, ste_pc))
```

#### Заключение
Проверялась гипотеза о равенстве средних рейтингов для платформ Xbox One и PC. Для проверки гипотезы исползовался тест Стьюдента с помощью метода <code>ttest_ind (array1, array2, equal_var)</code>, поскольку наши выборки относятся к разным генеральным совокупностям. Чтобы определить значения equal_var для этого метода использовался тест Барлетта. В итоге выяснилось, что дисперсии выборок различаются незначительно, а значит equal_var надо передать значение True. 

Средние рейтинги для Xbox One и PC составили 4.81±0.20 и 5.09±0.14 соответственно, p-value оказалось больше порогового значения $\alpha$ (0.24 и 0.05), а значит нулевую гипотезу нельзя отвергнуть.

### Сравнение пользовательких рейтингов для различных жанров

Необходимо проверить гипотезу о том, что пользовательские рейтинги для игр жанра "экшн" и "спортивные соревнования" различаются.  Сформулируем нулевую и альтернативную гипотезу следующим образом:
* H$_{0}$: средние пользовательские рейтинги для выбранных жанров одинаковые;
* H$_{1}$: средние пользовательские рейтинги для выбранных жанров отличаются;

Сравнение гипотез будем проводить так же, как и в предыдущем разделе

```python
# тест Барлетта
action_set = data_actual.query('genre=="Action"')
sport_set = data_actual.query('genre=="Sports"')
p = st.bartlett(action_set['user_score'], sport_set['user_score'])
print(np.where(p.pvalue < 0.05, 'Выборки значительно различаются: equal_var=False', 
         'Выборки различаются незначительно: equal_var=True'))
```

```python
result = st.ttest_ind(action_set['user_score'], sport_set['user_score'], equal_var=False)
alpha = 0.05
print('p-value: ', result.pvalue)
print(np.where(result.pvalue < alpha, 'Отвергаем нулевую гипотезу: пользовательские рейтинги отличаются', 
         'Принимаем нулевую гипотезу: рейтинги практически одинаковые'))

mean_action = action_set['user_score'].mean()
mean_sport = sport_set['user_score'].mean()
ste_action = np.std(action_set['user_score'], ddof=1)/np.sqrt(len(action_set['user_score']))
ste_sport = np.std(sport_set['user_score'], ddof=1)/np.sqrt(len(sport_set['user_score']))
print('Средний рейтинг для жанра "экшн": {0:.2f}\u00B1{1:.2f}'.format(mean_action, ste_action))
print('Средний рейтинг для жанра "спортивные соревнования": {0:.2f}\u00B1{1:.2f}'.format(mean_sport, ste_sport))
```

#### Заключение
Проверялась гипотеза о равенстве пользовательских рейтингов для игр различных жанров: экшн и спортивные соревнования. Для проверки гипотезы исползовался тест Стьюдента с помощью метода ttest_ind (array1, array2, equal_var), поскольку наши выборки относятся к разным генеральным совокупностям. Чтобы определить значения equal_var для этого метода использовался тест Барлетта. В итоге выяснилось, что дисперсии выборок различаются значительно, а значит equal_var надо передать значение False. 

Средние рейтинги для экшн-игр и игр-спортивных соревнований составили 3.53±0.09 и 4.12±0.21, p-value оказалось меньше порогового значения $\alpha$ (0.0006 и 0.05), а значит нулевая гипотеза должна быть отвергнута.
