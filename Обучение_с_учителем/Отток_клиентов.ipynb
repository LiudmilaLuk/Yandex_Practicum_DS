```python
data = pd.read_csv('/datasets/Churn.csv')

display(data.head())

print(data.shape)
display(data.describe())
print(data.info())
print('Число дубликатов: ', data.duplicated().sum())

plt.figure(figsize=(20, 16))
data.hist(bins=50, figsize=(20,15), color='seagreen', linewidth=2)
plt.grid()
plt.show()
```

В нашем датасете содержится 10000 объектов, для каждого из которых определены 14 признаков. Из них категориальными являются следующие: Geography, Gender, Tenure, NumOfProducts, HasCrCard, IsActiveMember и Exited. Некторые из них мы перекодируем в дальнейшем. Также в датасете есть столбцы, которые нам в дальнейшем не понадобятся (а при обучении модели еще и могут способствовать ее переобучению): Surname и UserID -- избавимся от этих столбцов. Столбец RowNumber избыточен, тк дублирует индексы, его мы тоже удалим. Столбец Tenure содержит пропущенные значения, нужно избавиться от них.

В распределениях кредитного рейтинга, возраста клиентов и баланса на счету заметны странные выбросы. Необходимо понять, чем они обусловлены и как их учитывать в дальнешем.

#### Что необходимо сделать:
* Заменить названия столбцов на более удобные.
* Удалить несколько столбцов: RowNumber, CustomerId и Surname -- они в дальнейшем нам не пригодятся.
* Заполнить пропуски в столбце Tenure, привести тип переменных в этом столбце к int.
* Посмотреть подробнее на распределения переменных Age, CreditScore и Balance.

Удалим лишние столбцы и переименуем остальные:

```python
data = data.drop(['RowNumber','CustomerId', 'Surname'], axis=1)
data = data.rename(columns={
    'CustomerId': 'customer_id',
    'CreditScore': 'credit_score',
    'NumOfProducts': 'num_of_products',
    'HasCrCard': 'has_cr_card',
    'IsActiveMember': 'is_active_member',
    'EstimatedSalary': 'estimated_salary'
})
data.columns= data.columns.str.lower()
data.columns
```

Заполним пропуски в столбце tenure. Посмотрим, есть ли какие-либо зависимости между столбцом tenure и другими колонками:

```python
display(data.corr()['tenure'])
print('Общее число пропусков: {0:0.1%}'.format(data['tenure'].isna().sum()/len(data['tenure'])))
```

Значимых корреляций нет ни с одним из столбцов, процент пропусков большой, но не значительный. Чтобы не искажать данные для обучения модели,избавимся от них:

```python
data.dropna(subset=['tenure'], inplace=True)

data['tenure'] = data['tenure'].astype(int)
display(data.head())
```

Посмотрим, как коррелируют между собой признаки:

```python
plt.figure(figsize=(20, 16))
s=sns.heatmap(data.corr())
s.set_title('Значения уровня корреляциий между признаками')
plt.show()
```

Признаки не сильно коррелируют между собой, исключение -- возраст клиентов и их статус (являются ли они клиентами или уже нет): возможно, эта корреляция обусловлена тем, что люди более старшего возраста перестают по тем или иным причинам пользоваться услугами банка (например, выходят на пенсию и не используют карточки банка).

#### Вывод:
Мы провели предобработку данных:
1. заменили названия столбцов на более удобные;
2. убрали лишние столбцы;
3. удалили строки с пропусками в столбце tenure;
4. проверили данные на наличие дубликатов.

Странный пик в последнем бине. Посмотрим, как распределяется рейтинг в зависимости от других признаков:

#### Исследование выбросов в распределении по возрастам

Распределение имеет несколько выбросов, равноудаленных друг от друга. Возможно, это как-то зависит от статуса пользователя: может быть, клиенты определенного возраста чаще покидают банк? Проверим корреляцию с признаком exited и другими.

```python
plt.figure(figsize=(20, 20))
a=sns.histplot(x='age', data=data, color='seagreen')

s=sns.histplot(x='age', data=data, hue='exited', fill=False, palette=['lime', 'red'], 
               linewidth=3, element='step')
s.set(title='Распределение возрастов клиентов', xlabel='Возраст, годы', ylabel='Число клиентов')
plt.grid()
plt.show()

print(data.corr()['age'])
```

Распределения клиентов как действующих, так и отказавшихся от услуг банка совпадают. Больше всего возраст клиентов коррелирует с признаками exited и is_active_member. Построим бокс-плоты для них:

```python
f, axes = plt.subplots(1, 2, figsize=(20, 20), sharex=False)

exited = sns.boxplot(x='exited', y='age', hue='exited', dodge=False, 
                data=data, ax=axes[0], palette=['seagreen', 'palegreen'])
exited.set(xlabel='Статус клиента', ylabel='Возраст клиента')
exited.grid()

exited.set_xticklabels(['Находится в клиентской базе','Отказалася от услуг'])
is_active = sns.boxplot(x='is_active_member', y='age', hue='is_active_member', dodge=False, 
                data=data, ax=axes[1], palette=['seagreen', 'palegreen'])
is_active.set(xlabel='Уровень активности клиента', ylabel='Возраст клиента')
is_active.grid()
is_active.set_xticklabels(['Не совершает операции','Совершает операции'])

exited.legend([],[], frameon=False)
is_active.legend([],[], frameon=False)

plt.show()
```

#### Вывод
От услуг банка в среднем чаще отказываются люди старше 45 лет. Возможно, это связано с какими-то особыми условиями получения услуг людей более младшего возраста (меньшая ставка по ипотеке для молодых семей, кредит на образование на более выгодных условиях и тп). В среднем после 37 лет клиенты банка менее активны. Однако это не объясняет странную структуру распределения возрастов клиентов, возможно, дополнительная информация об услугах банка и способах формирования данной выборки могла бы прояснить ситуацию.

## Создание моделей и выбор оптимальных параметров

В работе мы рассмотрим три модели:
* на основе дерева решений;
* на основе случайного леса;
* на основе линейной регресии.

Чтобы несколько раз не повторять одно и то же действие, создадим функцию, которая на вход будет принимать тип модели и список признаков, а оптимальные параметры модели (ищем максимальный скор для F1 на валидационной выборке).

```python
#------код модели:---------------------
# 0 -- дерево решений
# 1 -- случайный лес
# 2 -- линейная регрессия
#--------------------------------------

def objective_tree(trial: optuna.trial.Trial, fast_check=True):
    
    max_depth = trial.suggest_int('max_depth', 5, features_train.shape[1])
    min_samples_split = trial.suggest_int('min_samples_split', 2, 20)
    min_samples_leaf = trial.suggest_int('min_samples_leaf', 2, 20)
    model = DecisionTreeClassifier( max_depth=max_depth, min_samples_split=min_samples_split, 
                                   min_samples_leaf=min_samples_leaf, class_weight = is_weighted,
                                   random_state=299792458)
    
    model.fit(features_train, target_train)
    return f1_score(target_valid, model.predict(features_valid))

def objective_forest(trial: optuna.trial.Trial, fast_check=True):
    
    max_depth = trial.suggest_int('max_depth', 2, 32)
    max_est = trial.suggest_int('max_est', 2, 100)
    model = RandomForestClassifier(max_depth=max_depth, n_estimators=max_est, class_weight = is_weighted,
                                   random_state=299792458, n_jobs=-1)
    
    model.fit(features_train, target_train)
    return f1_score(target_valid, model.predict(features_valid))

def objective_regression(trial: optuna.trial.Trial, fast_check=True):
    
    regularization = trial.suggest_float('regularization', 1e-10, 1e10, log=True)
    tol = trial.suggest_float('tol', 1e-4, 1e4, log=True)
    max_iter = trial.suggest_int('max_iter', 10, 1e5)
    
    penalty = trial.suggest_categorical('penalty', ['l1', 'l2'])
    
    if penalty == 'l1':
        solver = 'saga'
    if penalty == 'l2':
        solver='lbfgs'
    
    model = LogisticRegression(penalty=penalty, solver=solver,
                                        C=regularization,  tol=tol, max_iter=max_iter, class_weight = is_weighted,
                               random_state=299792458, n_jobs = -1)
    model.fit(features_train, target_train)
    
    return f1_score(target_valid, model.predict(features_valid))

def model_runner(classifier_type, features_train, target_train, features_valid, target_valid, n_trials):
    types = ['Tree', 'Random Forest', 'Linear regression']
    
    
    features_train = features_train
    features_valid = features_valid
    target_train = target_train
    target_valid = target_valid
    
    optuna.logging.set_verbosity(optuna.logging.WARNING)
    study = optuna.create_study(direction='maximize')
    
    if classifier_type == 0:
        study.optimize(objective_tree, n_trials = n_trials)
    if classifier_type == 1:
        study.optimize(objective_forest, n_trials = n_trials)
    if classifier_type == 2:
        study.optimize(objective_regression, n_trials = n_trials)
    types = ['Decision tree model', 'Random forest model', 'Logistic regression model']
    print('*********************{0}*********************'.format(types[classifier_type]))
    
    params_list  = []

    for key, value in study.best_params.items():
        print(f'{key}: {value}')
        params_list.append(value)
        
    print('F1:',study.best_value)
    
    return  params_list
```

Разобьем исходные данные на три части в соотношении 3:1:1. Для этого сначала разобьем исходный датасет на обучающую и валидационную выборки, а потом разделим пополам валидационную выборку.

```python
data_train, data_valid = train_test_split(data_clear, test_size=0.4, random_state=299792458)
data_valid, data_test = train_test_split(data_valid, test_size=0.5, random_state=299792458)
print(data_train.shape, data_valid.shape, data_test.shape)

data_train = data_train.reset_index(drop=True)
data_valid = data_train.reset_index(drop=True)
data_test = data_train.reset_index(drop=True)
```

Отмасштабируем численные признаки (столбцы balance, estimated_salary, credit_score) в тренировочной выборке.

```python
scaler = StandardScaler() 
scaler.fit(data_train[['balance', 'credit_score', 'estimated_salary']])
scaled = pd.DataFrame(scaler.transform(data_train[['balance', 'credit_score', 'estimated_salary']]), columns=[
    'balance_scaled', 'credit_score_scaled', 'estimated_salary_scaled'])
data_train.loc[:, 'balance'] = scaled['balance_scaled']
data_train.loc[:, 'credit_score'] = scaled['credit_score_scaled']
data_train.loc[:, 'estimated_salary'] = scaled['estimated_salary_scaled']

```

Теперь выберем модель, которая лучше всего себя покажет на несбалансированной выборке. Для определения качества моделей воспользуемся F1-метрикой и AUC-ROC.

### Проверка моделей на несблансированных данных

Проверим, какая из трех моделей (на основе решающего дерева, случайного леса или логистическая регрессия) покажут лучшее значение F1-score. Лучшая модель должна обладать значением полноты F1 >= 0.59.

```python
import warnings
warnings.filterwarnings('ignore')

features_train = data_train.drop('exited', axis=1)
target_train = data_train['exited']

features_valid = data_valid.drop('exited', axis=1)
target_valid = data_valid['exited']

features_test = data_test.drop('exited', axis=1)
target_test = data_test['exited']

# мы не учитываем дисбаланс, поэтому передадим следуещее значение для аргумента class_weight
is_weighted = None

tree_params = model_runner(0, features_train, target_train, features_valid, target_valid, 300)
forest_params = model_runner(1, features_train, target_train, features_valid, target_valid, 300)
rgr_params = model_runner(2, features_train, target_train, features_valid, target_valid, 400)
```

### Тестирование модели на несбалансированных данных

 Модели на основе случайного леса и решающего дерева показали лучшее значение метрики F1, проверим их на тестовой выборке, а затем лучшую модель -- на адекватность.

```python
model_tree = DecisionTreeClassifier( max_depth=tree_params[0], min_samples_split=tree_params[1], 
                                   min_samples_leaf=tree_params[2], random_state=299792458).fit(
    features_train, target_train)
model_forest = RandomForestClassifier(max_depth=forest_params[0], n_estimators=forest_params[1],
                                      random_state=299792458, n_jobs=-1).fit(
    features_train, target_train)

predictions_tree = model_tree.predict(features_test)
f1_tree = f1_score(target_test, predictions_tree)

predictions_forest = model_forest.predict(features_test)
f1_forest = f1_score(target_test,predictions_forest)

print('F1 для решающего дерева: {0};\nF1 для случайного леса: {1}'.format(f1_tree, f1_forest))
```

На тестовой выборке скор у модели на основе решающего дерева выше. Проверим модель на адекватность.

Сначала убедимся, что модель предсказывает нам только 0 и 1, и выведем значение всех метрик качества:

Наша модель верно относит клиента к классу exited = 1 только в 48% всех случаев, при этом значение precision достаточно высоко: это может быть связано с дисбалансом классов в наших данных -- клиентов отказавшихся от услуг банка более чем в три раза меньше, чем клиентов, которые не разрывали отношений с банком.

Теперь сравним матрицу ошибок нашей модели с матрицей ошибок для модели, которая бы предсказывала только часто встречающийся класс (majority), и генератором, который предсказывал бы один из классов с равной вероятностью(uniform). Построим такие модели с помощью генератора DummyClassifier.

```python
model_majority = DummyClassifier(strategy='most_frequent')
model_majority.fit(features_train, target_train) 
predictions_majority = model_majority.predict(target_test) 


model_uniform = DummyClassifier(strategy='uniform')
model_uniform.fit(features_train, target_train) 
predictions_uniform = model_uniform.predict(target_test) 


matrix_tree = confusion_matrix(target_test, predictions_forest)
print('Матрица ошибок (Решающее дерево):\n', matrix_tree)

matrix_majority = confusion_matrix(target_test, predictions_majority)
print('Матрица ошибок (majority):\n', matrix_majority)

matrix_uniform = confusion_matrix(target_test, predictions_uniform)
print('Матрица ошибок (uniform):\n', matrix_uniform)
```

Наша модель отличается от моделей на основе простейших генераторов. Построим для этой модели ROC-кривую и оценим площадь под ней.

```python
probabilities = model_tree.predict_proba(features_test)
probabilities_one = probabilities[:,1]
fpr, tpr, thresholds = roc_curve(target_test, probabilities_one)
auc_roc = roc_auc_score(target_test, probabilities_one)

plt.subplots(1, figsize=(10,10))
s = sns.lineplot(x=fpr, y=tpr, linewidth=5, color='seagreen', label='ROC-кривая для модели Decision Tree')
constant = sns.lineplot(x=[0,1], y=[0,1], ls='--',  linewidth=5, color='lime', label='Случайная модель')
s.set(title='Cравнение ROC-кривых для модели решающего дерева и случайной модели', 
      xlabel='False Positive Rate', ylabel='True Positive Rate' )
s.grid()
plt.legend(loc='lower right')
plt.show()

print('AUC-ROC =', auc_roc)
```

## Тестирование модели

Лучшее значение F1 показала модель на основе случайного леса. Проверим ее на тестовой выборке.

```python
model_forest = RandomForestClassifier(max_depth=forest_params_weighted[0], n_estimators=forest_params_weighted[1], 
      class_weight='balanced',random_state=299792458, n_jobs=-1).fit(features_train, target_train)
predictions = model_forest.predict(features_test)

print('F1-score: {0:.2}'.format(f1_score(target_test, predictions)))
print(np.unique(predictions))
```

```python
model_majority = DummyClassifier(strategy='most_frequent')
model_majority.fit(features_train, target_train) 
predictions_majority = model_majority.predict(target_test) 


model_uniform = DummyClassifier(strategy='uniform')
model_uniform.fit(features_train, target_train) 
predictions_uniform = model_uniform.predict(target_test) 


matrix_forest = confusion_matrix(target_test, predictions)
print('Матрица ошибок (Случайный лес):\n', matrix_forest)

matrix_majority = confusion_matrix(target_test, predictions_majority)
print('Матрица ошибок (majority):\n', matrix_majority)

matrix_uniform = confusion_matrix(target_test, predictions_uniform)
print('Матрица ошибок (uniform):\n', matrix_uniform)
```

Предсказания нашей модели отличаются от предсказаний случайной и константной модели.

```python
probabilities = model_forest.predict_proba(features_test)
probabilities_one = probabilities[:,1]

fpr, tpr, thresholds = roc_curve(target_test, probabilities_one)
auc_roc = roc_auc_score(target_test, probabilities_one)

plt.subplots(1, figsize=(10,10))

plt.plot(fpr, tpr, linewidth=5, color='seagreen')
plt.title('Cравнение ROC-кривых для модели случайного леса и случайной модели')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.plot([0,1], [0,1], ls='--',  linewidth=5, color='lime', label='Случайная модель')
plt.grid()
plt.show()


print('AUC-ROC = {0:.2}'.format(auc_roc))
```

Значение AUC-ROC для нашей модели выше чем для случайной модели. Значение метрики F1 = 0.62. Модель на основе случайного леса прошла тестирование.
